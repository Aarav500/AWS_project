```python

mport pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter

# Sample DataFrame with Age and Survival columns
data = {'Age': [22, 35, 45, 50, 60, 70, 75, 80],
        'Survival': [1, 1, 0, 1, 0, 1, 1, 0]}  # 1: Survival, 0: Non-survival
df_survival = pd.DataFrame(data)

# Create Kaplan-Meier estimator
kmf = KaplanMeierFitter()

# Fit the survival data
kmf.fit(durations=df_survival['Age'], event_observed=df_survival['Survival'])

# Plot the survival function
kmf.plot_survival_function(show_censors=True)
plt.title('Survival Distribution by Age')
plt.xlabel('Age')
plt.ylabel('Survival Probability')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create a new column 'fare_category' based on fare ranges
fare_bins = [0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500]
fare_labels = ['0-50', '51-100', '101-150', '151-200', '201-250', '251-300', '301-350', '351-400', '401-450', '451-500', '500+']
titanic_data['fare_category'] = pd.cut(titanic_data['fare'], bins=fare_bins, labels=fare_labels, right=False)

# Calculate survival rate for each fare category
fare_survival_rate = titanic_data.groupby('fare_category')['survived'].mean()

# Create a box plot
plt.figure(figsize=(12, 8))
sns.boxplot(x='fare_category', y='fare', data=titanic_data, hue='survived', showfliers=False)
plt.title('Fare vs Survival Rate in Titanic')
plt.xlabel('Fare Category')
plt.ylabel('Fare')
plt.show()
```
